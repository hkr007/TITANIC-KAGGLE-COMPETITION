```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv("titanic_train.csv")
```

```python
df
```

## removing unwanted features 
## Feature Selection 
## i am assuming that if i remove these features i will get model good performance 

## target variable is Pclass here 

```python
df.drop(['PassengerId','Name','Ticket','Embarked','Cabin','SibSp'],axis=1,inplace=True)

#my assumtions is these feature not need to build a model 
##so i am removing 
```

```python
df.head()
```

## Now will start Data Preprocessing Activity 

```python
df.isnull().sum()
```

```python
#see age have null values,so we need to use imputation methods
df['Age'].describe()
```

```python
df['Age'].fillna(df['Age'].mean(),inplace=True)
```

```python
df.isnull().sum()
```

```python
#we have categorical values in sex featue....so need to use encoding techniques 
#by using get dummies will do 
dummies=pd.get_dummies(df['Sex'],drop_first=True)#get dummyt and remove one first column 
```

```python
df1=pd.concat([df,dummies],axis=1)
```

```python
df1
```

```python
df=df1.drop(['Sex'],axis=1)
```

## use standard Scalar----to perform best --reduse values range root, log...etc
## scaling down 

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
```

```python
features=['Age','Fare']
df[features]=sc.fit_transform(df[features])
```

```python
df
#now scaled down ----cal
```

```python
x=df.drop(['Survived'],axis=1)
```

```python
y=df['Survived']
```

```python
y
```

```python
from sklearn.tree import DecisionTreeClassifier 
from sklearn.neighbors import KNeighborsClassifier 
from sklearn.svm import SVC
from sklearn.model_selection import GridSearchCV
```

## now will create pipeline to check all models ----performance

```python
#parameters ----and checking all model perameters---
model_parms={
    'DecisionTreeClassifier':{
        'model':DecisionTreeClassifier(),
        'param':{
            'criterion':['gini','entropy']
        }
        
        
    },
    'KNeighborsClassifier':{
        'model':KNeighborsClassifier(),
        'param':{
            'n_neighbors':[5,10,15,20,25]
        }
        
        
    }
        
    }
    
    
```

```python
scores=[]
for model_name,mp in model_parms.items():
    model_selection=GridSearchCV(estimator=mp['model'],param_grid=mp['param'],cv=5,return_train_score=False)
    model_selection.fit(x,y)
    scores.append(
        {
            'model':model_name,
            'best_score':model_selection.best_score_,
            'best_params':model_selection.best_params_
        
        
        
    }
    )
```

```python
scores
```

# we can compare more models like this and fit best model to test the data

```python
#Now will fit SVC model 
model_svc=SVC(C=100,kernel='rbf')
```

```python
model_svc.fit(x,y)
```
